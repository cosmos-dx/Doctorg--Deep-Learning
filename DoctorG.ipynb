# **Preprocessing**

```python
# 1. remove code col
# 2. lowercase disease name
# 3. symptom lowercase
# 4. symptom tokeinization
# 5. symptom stop word removal
# 6. description lowercase
# 7. description remove symbols
# 8. description remove extra whitespace
# 9. weight scaling
# 10. create embeddings of symptoms
import pandas as pd
df = pd.read_csv('doctorg_data.csv')
df.head()
```

```python
df.drop('code', axis=1, inplace=True)
df.head()
```

```python
df[['name', 'symptom']] = df[['name', 'symptom']].apply(lambda x: x.str.lower())
df.head()
```

```python
import spacy
nlp =  spacy.load("en_core_web_sm")

def process_symptom(text):
  doc = nlp(str(text).lower())
  tokens = [token.text for token in doc if not token.is_stop and token.is_alpha]
  return tokens

df['symptom'] = df['symptom'].apply(process_symptom)
df.head()
```

```python
import re
def cleared_description(desc):
  doc = nlp(str(desc).lower())
  cleared_desc = re.sub(r'["{}[\]]', '', desc)
  cleared_desc_fin = re.sub(r'\s+', ' ', cleared_desc)
  return cleared_desc_fin.strip()

df['description'] = df['description'].apply(cleared_description)
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['weight'] = scaler.fit_transform(df[['weight']])
df.head()
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
df['embeddings'] = df['symptom'].apply(lambda x: model.encode(' '.join(x)))
df.head()
```

```python
df.to_csv('doctorg_processed.csv', index=False)
```

# **Model Testing**

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.preprocessing import LabelEncoder
from sentence_transformers import SentenceTransformerModelCardData

df = pd.read_csv('doctorg_processed.csv')

label_encoder = LabelEncoder()
df["disease_encoded"] = label_encoder.fit_transform(df["name"])

X = np.vstack(df.apply(lambda row: np.fromstring(row["embeddings"].strip("[]"), sep=' ') * row["weight"], axis=1).values)
y = tf.keras.utils.to_categorical(df["disease_encoded"], num_classes=len(label_encoder.classes_))

model = Sequential([
    Dense(128, activation="relu", input_shape=(X.shape[1],)),
    Dropout(0.3),
    Dense(64, activation="relu"),
    Dropout(0.3),
    Dense(len(label_encoder.classes_), activation="softmax")
])

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
model.fit(X, y, epochs=50, batch_size=16, validation_split=0.2)

model.save("disease_prediction_model.h5")
```

```python
model = load_model("disease_prediction_model.h5")

df = pd.read_csv("doctorg_processed.csv")
label_encoder = LabelEncoder()
df["disease_encoded"] = label_encoder.fit_transform(df["name"])

GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
genai.configure(api_key=GOOGLE_API_KEY)
model_llm = genai.GenerativeModel("models/gemini-2.0-flash-exp")

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
llm_my = LLM(model="gemini/gemini-2.0-flash-exp", temperature=0.7)

#Disease Information Specialist
disease_information_specialist = Agent(
    role="Disease Information Specialist",
    goal="To provide users with calm, informative, and helpful guidance regarding their described symptoms, suggest general remedies, advise on next steps, and categorize the potential disease.",
    backstory="You are a compassionate and knowledgeable AI agent specializing in providing initial guidance on health concerns. You understand that users might be feeling anxious and your primary goal is to offer clear, easy-to-understand information and help them make informed decisions about their health."
    ,
    llm = llm_my,
)
def create_disease_info_task(disease_description, top_disease):
    return Task(
        description=f"""Based on your mentioned symptoms, a potential condition is '{top_disease}'. Here is a brief description:

        {disease_description}

        Your objectives are to:
        1. Gently reassure the user regarding the possibility of '{top_disease}'.
        2. Suggest general, widely accepted home remedies or self-care tips for the symptoms described. Emphasize that these are not substitutes for professional medical advice.
        3. Provide general information about '{top_disease}' in simple terms, including common symptoms and potential causes (based on the provided description).
        4. Guide the user on appropriate next steps, including when to consult a doctor or seek professional medical help for proper diagnosis and treatment.
        5. Categorize '{top_disease}' into a broad category.

        Please provide a concise and easy-to-understand response addressing these points in a helpful and reassuring manner.""",
        agent=disease_information_specialist,
        expected_output="A helpful and informative response about the potential condition, including reassurance, general remedies, information, next steps, and categorization."
    )

def extract_symptoms(natural_language_input):
    """
    Extracts a list of symptoms mentioned in the input and removed stopwords using spacy.
    """
    prompt = f"""Please extract the symptoms mentioned in the following text and return them as a comma-separated list:

    \"{natural_language_input}\"

    Symptoms:"""
    response = model_llm.generate_content(prompt)
    symptom_string = response.text.strip()
    symptoms = [s.strip() for s in symptom_string.split(',')]
    processed_symptoms = []
    for symptom in symptoms:
      doc = nlp(symptom.lower())
      filtered_tokens = [token.text for token in doc if not token.is_stop and not token.is_punct]
      processed_symptoms.append(" ".join(filtered_tokens).strip())
    return [s for s in processed_symptoms if s]

def predict_disease(user_symptoms):
    sentence_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
    user_embedding = sentence_model.encode(" ".join(user_symptoms)).reshape(1, -1)

    predictions = model.predict(user_embedding)[0]
    top_5_indices = np.argsort(predictions)[-5:][::-1]
    top_5_diseases = [label_encoder.classes_[i] for i in top_5_indices]
    top_5_probs = [predictions[i] for i in top_5_indices]

    return list(zip(top_5_diseases, top_5_probs))

def get_disease_description(disease_name, dataframe):
    """
    Retrieves the description of a given disease from the DataFrame.
    """
    disease_row = dataframe[dataframe['name'] == disease_name]
    if not disease_row.empty:
        return disease_row['description'].iloc[0]
    else:
        return f"Description for '{disease_name}' not found."

# User input (replace with actual user input if needed)
# user_input ="I am feeling depressed and anxious and bit nervousness and sometimes insomania and palpitations and little bit of dizzy and breathing fast, irregular heartbeat, depressive psycotic symptoms and shortness of breath and dizziness"
user_input = " I am having hoarse voice, sore throat, difficulty in speaking, cough, throat swelling, skin swelling, and retention in urine."
# user_input = "I am having fatty liver, Constipation,  death"
symptom_arr = extract_symptoms(user_input)
print("Extracted Symptoms:", symptom_arr)

predicted_diseases = predict_disease(symptom_arr)

print("\nBased on these symptoms, the top 5 most likely diseases are:")
if predicted_diseases:
    for i, (disease, probability) in enumerate(predicted_diseases):
        confidence_percentage = probability * 100
        print(f"- {disease} with a probability of {confidence_percentage:.2f}%")

    top_disease = predicted_diseases[0][0]
    description = get_disease_description(top_disease, df)

    disease_info_task = create_disease_info_task(description, top_disease)

    crew = Crew(
        agents=[disease_information_specialist],
        tasks=[disease_info_task],
        verbose=False
    )

    print("\n--- AI Assistant Guidance ---")
    result = crew.kickoff()
    print(result)
else:
    print("Could not predict any disease based on the provided symptoms.")

```
